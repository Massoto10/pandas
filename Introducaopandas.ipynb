```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
alugueis.query('@imoveis_comerciais in Tipo')

# "query(...)" o query serve para selecionar tais dados dentro de uma coluna. nesse caso ele teve que selecionar os dados da minha lista de imoveis_comerciais na coluna Tipo

# o "@" amtes do "imovies_comerciais" diz que estamos referenciando essa lista mesmo que dentro de parenteses
```

```python
alugueis.query('@imoveis_comerciais not in Tipo')
```

```python
alugueis_residenciais = alugueis_residenciais.query('Tipo == "Apartamento"')
alugueis_residenciais.head()
```

```python
alugueis_residenciais.fillna(0)

# fillna: substitui os valores nulos por um valor especifico entre parenteses
```

```python
alugueis_residenciais = alugueis_residenciais.fillna(0)
```

```python
alugueis_residenciais.query('Valor == 0 | Condominio == 0').index

#index: extrais das linhas selecionadas apenas o seu index
```

```python
alugueis_residenciais_inconsistentes = alugueis_residenciais.query('Valor == 0 | Condominio == 0').index
```

```python
alugueis_residenciais.drop(alugueis_residenciais_inconsistentes,axis = 0, inplace=True)

# drop: Apapa os caracteres
# axis: ecolhe apagar entre: 0 - Linha e 1 - colunas
# inplace: Já aplica o valor obtido sem precisar atribuir novamente
```

```python
alugueis_residenciais.drop('Tipo', axis=1, inplace= True)
```

```python
alugueis = pd.read_csv('alugueldados.csv', sep = ';')
```

```python
alugueis = alugueis.fillna(0)
```

```python
alugueis['Valor_por_mes'] = alugueis['Valor'] + alugueis['Condominio']
alugueis['Valor_por_ano'] = alugueis['Valor_por_mes']*12 + alugueis['IPTU']

# para fazer a criação de novas colunas simplesmente referência ela e escreve a lógica por traz da coluna em si
```

```python
alugueis
```

```python
#essa coluna deve possuir uma sumarização das principais informações dos imóveis que serão apresentadas no site: tipo de imóvel, bairro, quantidade de quartos e vagas de garagem;

alugueis['Descricao'] = alugueis['Tipo'] + ' em ' + alugueis['Bairro'] + ' com ' + alugueis['Quartos'].astype(str) + ' quartos e ' + alugueis['Vagas'].astype(str) + ' vagas de garagem'

alugueis

# astype(str): Diz ao programa, encare esse dado com string
```

```python
alugueis['Possui_suite'] = alugueis['Suites'].apply(lambda x : 'Sim' if x > 0 else 'Não')

alugueis

# para fazer a lógica ultilizamos função lambda
```
