Para fazer esse agrupamento de forma correta, precisamos alterar o formato do DataFrame. A nossa tabela possui as informações de emissão separada por colunas de anos, mas precisamos alterar o formato para:

Apenas uma coluna contendo as informações dos anos. Assim, cada linha vai conter uma informação de um ano diferente;
Apenas uma coluna com os dados de emissão.

```python
emissoes_gases.loc[:,'Nível 1 - Setor':'Produto'].columns

# o loc e o query tem funções muito semelhantes, nesse caso o loc está selecionando todas as linhas, uma vez que n tem nenhuma filtragem antes dos ":" e além disso seleciona todas as colunas entre nivel setro e produto.

# o papel do columns é fazer que seja selecionado somente as colunas compreendidas nesse intervalo
```

```python
colunas_info = list(emissoes_gases.loc[:,'Nível 1 - Setor':'Produto'].columns)
```

```python
emissoes_gases.loc[:,1970 : 2021].columns
```

```python
colunas_anos = list(emissoes_gases.loc[:,1970 : 2021].columns)
```

```python
emissoes_gases.melt(id_vars=colunas_info, value_vars=colunas_anos, var_name= 'Ano', value_name= 'Emissão')

# melt: serve para desempilhar colunas, trasnformar colunas em linhas
# id_vars: fixam as colunas que serviram de id, ou seja que se manterão como colunas
# value_vars: os valores que serão transofrmados em linhas
# var_name: nome da nova coluna formada pelas colunas "derretidas"
# value_name: nome da coluna criada pelos valores que as colunas "derretidas" carregavam
```

```python
emissoes_gases_anos = emissoes_gases.melt(id_vars=colunas_info, value_vars=colunas_anos, var_name= 'Ano', value_name= 'Emissão')

# melt "derrete" uma coluna em duas, nessas duas. Uma abrigará o valor da
# coluna e a outra o seu respectivo valor.
```

Na nossa tabela, temos uma coluna intitulada "Gás" que identifica qual gás emitido. Além disso, como realizamos aquela transformação nos dados, existe uma única coluna contendo a quantidade de emissão de cada um dos registros.

Como cada uma das linhas nos informa uma emissão de uma indústria para cada um dos anos, vamos ter informações duplicadas de cada gás - já que as indústrias emitem os mesmos tipos de gases.

Para obter a nossa resposta, precisamos somar a emissão para cada tipo de gás. Isto é, somar a coluna "Emissão" com base nos tipos de gás da coluna "Gás".



```python
emissoes_gases_anos.groupby('Gás')
```

```python
emissoes_gases_anos.groupby('Gás').get_group('CO2 (t)')
# o get_group está especificando o tipo informação a ser filtrada na tabela a baixo
```

```python
emissoes_gases_anos.groupby('Gás').sum(numeric_only=True).sort_values('Emissão', ascending=False)
# agrupo pela coluna 'Gás' e somo apenas os numericos, coloco em ordem dentro
# da coluna 'Emissão' mas essa ordem é descrescente (ascending = False)
```

```python
populacao_estados = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Pandas: Seleção e Agrupamento de dados/POP2022_Municipios.xls', header = 1, skipfooter=34)
populacao_estados.head()

# Li o arquivo em excel, Header diz por onde começar, (padrão == 0).
# skipfooter = 34 diz que devo pular as últimas 34 linhas da minha planilha
# sendo assim eu ignoro as últimas 41 linhas da minha planilha
```

```python
populacao_estados.groupby('UF').sum(numeric_only=True)
# coloquei para somar somente os numéricos e a 'POPULAÇÃO' não foi somada
```

```python
populacao_estados[populacao_estados['POPULAÇÃO'].str.contains('\(', na = False)]
# após estudar a planilha percebo que dentro de populção possuo caracteres que
# me fazem ter uma string como as parenteses e o ponto
```

```python
populacao_estados = populacao_estados . assign(populacao_sem_parenteses = populacao_estados['POPULAÇÃO'] . replace('\(\d{1,2}\)' , '' , regex = True) ,
populacao = lambda x : x . loc[:, 'populacao_sem_parenteses'] . replace('\.', '', regex = True))

# assign tem o papel de criar uma nova coluna, na qual se chama 'populacao_sem_parenteses', na qual herda da coluna 'POPULAÇÃO' e que possui um replace que diz: ('\(\d{1,2}\)' , '' , regex = True) significa que o carcter de escape (\) está colocando parentêses que tenham dentro um ou dois digitos(d). buscam esse padrão (regex faz interpretar o padrão) e substitui por string vazia ''.

# criamos agora a coluna populacao e temos uma função lambda talque x : x (dentro da própria coluna de populção) que todas as linhas da coluna sem parenteses também vai passar por um replace com o caracter ponto(.) que também faz com que se caracterize como string toda a coluna.
```

```python
populacao_estados[populacao_estados['POPULAÇÃO'].str.contains('\(', na = False)]

# dessa maneiro eu pego as colunas que tinham os parenteses, para ver se foram tratados de forma correta e ainda me previso de possíveis erros com valores nulos (na = FALSE)
```

```python
populacao_estados = populacao_estados.astype({'populacao' : 'int64'})
# agora sim, consigo transformar a coluna de população em inteiro
```

```python
populacao_estados = populacao_estados.groupby('UF')[['populacao']].sum()

populacao_estados
```

```python
emissao_estados = emissoes_gases_anos[emissoes_gases_anos['Ano'] == 2021].groupby('Estado')[['Emissão']].sum().reset_index()

emissao_estados

# nesse caso queremos apenas as informações refente ao ano de 2021. de tal forma a agrupar os estados e realizar a soma das emissões. dessa forma também ultilizamos o reset_index() para que o 'Estado' não seja posto com index desse df
```

```python
dados_agrupados = pd.merge(emissao_estados, populacao_estados, left_on = 'Estado', right_on = 'UF')

dados_agrupados

#pd.merge tem a função de mesclar dois dfs. E ainda correlaciona-los com o left_on e o right_on
```

```python
dados_agrupados = dados_agrupados.assign(emissao_per_capita = (dados_agrupados['Emissão']/dados_agrupados['populacao']).round(2)).sort_values('emissao_per_capita', ascending = False)
dados_agrupados

# assign novamente criando mais uma coluna, nesse caso emissao_per_capita que consite na divisão de duas outras colunas do mesmo df, e arredondamos(round) para 2 casas decimais e ordenamos em ordem decrescente(ascending = False)
```
