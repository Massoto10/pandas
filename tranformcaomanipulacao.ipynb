```python
dados = pd.json_normalize(dados['info_moveis'])
dados

# nesse caso precisamos usar o normalize uma vez que os dados estavam contido dentro de um dicionário
```

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[3:])
dados

# com lista criada acima com os nomes das colunas, eu vejo as que possuem listas como conteúdo, pego sua posição e as "explodo" para conseguir separar os itens
```

```python
dados.reset_index(inplace = True, drop = True)
dados

# uso o reset_index para reiniciar o indice do data frame, uma vez que ficou desorganizado por conta da tratativa das listas.
# o inplace indica que quero modificar o meu data frame atual e essa modificação fica por responsabilidade do drop que tem como objetivo descartar o index original.
```

```python
dados.info()
```

```python
import numpy as np
```

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
# converto minha coluna "max_hospedes" para um inteiro
```

```python
dados.info()
```

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
# nas ultimas colunas realizei as trativas de dados necessárias para a manipulação das mesmas, transformando-as em inteiros e float
```

```python
dados['preco'] = dados['preco'].apply(lambda x : x.replace('$','').replace(',','').strip())

# aqui precisei ir além e tratar os dados da coluna preco para que fossem removidos os cifrões, as virgulas e os espaços em branco(strip)
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
# para agora sim conseguir realizar a conversão da string em float
```

```python
dados.info()
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x : x .replace('$','').replace(',',''))

dados

# nesse caso faço a tratativa de mais duas colunas(applymap) novamente removendo cifrão e a vírgula. nesse caso faço o uso da função lambda para percorrer item a item.
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)


# para agora sim poder converter a string em float
```

```python
dados
```

Tokenização: separar texto em partes menores
             "separar" "texto" "em" "partes" "menores"

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
dados
```

```python
dados['descricao_local']
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-z0-9\-\']',' ',regex = True)

dados['descricao_local']

# [^a-zA-Z0-9\-\']
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ',regex = True)

# (?<!\w)-(?!\w): lookbehind negativo e lookahead negativo
# (?<=\w)-(?=\w): lookbehind positivo e lookahead positivo
```

um pouco mais sobre regex: https://www.alura.com.br/artigos/principais-casos-uso-regex-para-tratamento-dados

```python
dados['descricao_local'] = dados['descricao_local'].str.split()

dados
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower().replace('[^a-z0-9\-\']',' ',regex = True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<=\w)-(?=\w)', '', regex = True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
dados
```
