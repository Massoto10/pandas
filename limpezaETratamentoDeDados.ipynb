```python
import seaborn as sns
```

```python
sns.boxplot(x=df_sem_nulo['cliente.tempo_servico'])

# aqui eu consgio obter o gráfico para entender melhor a situação dos meus dados qunto ao outlier
```

```python
q1 = df_sem_nulo['cliente.tempo_servico'].quantile(.25)
q3 = df_sem_nulo['cliente.tempo_servico'].quantile(.75)
```

```python
IQR = q3 - q1

limite_inferior = q1 - 1.5*IQR
limite_superior = q3 + 1.5*IQR

# nas duas células anteriores me preparei para usar as fórmulas presentes na foto entre os limites e o IIQ
```

```python
outliers = ((df_sem_nulo['cliente.tempo_servico'] < limite_inferior) | (df_sem_nulo['cliente.tempo_servico'] > limite_superior))

outliers
```

```python
df_sem_nulo[outliers]['cliente.tempo_servico']

# isso me retornar apeans as linhas que carregam o True
```

```python
df_sem_out = df_sem_nulo.copy()
```

```python
df_sem_out.loc[outliers, 'cliente.tempo_servico'] = np.ceil(
    df_sem_out.loc[outliers, 'conta.cobranca.Total'] / df_sem_out.loc[outliers, 'conta.cobranca.mensal']
)

# corrijo parte dos dados que estavam como outliers
```

```python
sns.boxplot(x = df_sem_out['cliente.tempo_servico'])
```

```python
q1 = df_sem_out['cliente.tempo_servico'].quantile(.25)
q3 = df_sem_out['cliente.tempo_servico'].quantile(.75)

IQR = q3 - q1

limite_inferior = q1 - 1.5*IQR
limite_superior = q3 + 1.5*IQR

outliers = ((df_sem_out['cliente.tempo_servico'] < limite_inferior) | (df_sem_out['cliente.tempo_servico'] > limite_superior))

df_sem_out[outliers]['cliente.tempo_servico']

# faço novamente os cálculos para verificar quais valores, mesmo depois da correção, continuam tão descrepantes
```

```python
df_sem_out = df_sem_out[~outliers]
df_sem_out

# dessa forma preciso retira-los do meu DF e por isso peço para não incluí=los
```

```python
sns.boxplot(x = df_sem_out['cliente.tempo_servico'])

# dessa forma, tenho um gráfioc bem mais equilibrado e dentro de uma margem de tolerância
```

```python
df_sem_out.reset_index(drop=True, inplace=True)
df_sem_out
```
